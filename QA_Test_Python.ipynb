Q1 — Count Payment Failures

ABC stores transaction logs. Given a list of status values, count how many failed transactions occurred.

Input:
["success","failed","success","failed","failed","pending"]

Output:
3

```python
transactions = ["success","failed","success","failed","failed","pending"]

failed_count = 0
for status in transactions:
    if status == "failed":
        failed_count += 1

print(failed_count)
```

Q2 — Validate Email Format (Basic)

Write a function to verify if an email is valid.

Rules:

Must contain @

Must contain .

@ must come before .

Example
test@abc.com  → True
invalid.abc@  → False


```python
def is_valid_email(email):

    # Rule 1: Must contain @
    if '@' not in email:
        return False

    # Rule 2: Must contain .
    if '.' not in email:
        return False

    # Get the index of @ and .
    at_index = email.find('@')
    dot_index = email.find('.')

    # Rule 3: @ must come before .
    if at_index >= dot_index:
        return False

    return True

# Example Test Cases:
print(f"test@abc.com  → {is_valid_email('test@abc.com')}")
print(f"invalid.abc@  → {is_valid_email('invalid.abc@')}")
```

Q3 — Find Duplicate Transaction IDs

Given a list of transaction IDs, return the duplicates.

Input:
["TX1001", "TX1002", "TX1001", "TX1003", "TX1002"]

Output:
["TX1001", "TX1002"]


```python
def find_duplicates(transactions):
    seen = set()
    duplicates = set()
    for tx_id in transactions:
        if tx_id in seen:
            duplicates.add(tx_id)
        else:
            seen.add(tx_id)
    return list(duplicates)

# Input:
transaction_ids = ["TX1001", "TX1002", "TX1001", "TX1003", "TX1002"]

# Output:
result = find_duplicates(transaction_ids)
print(result)
```

Q4 — API Response Validator

You get a list of API responses. Mark each as PASS if "status":200 else FAIL

Input:
responses = [200, 404, 200, 500]

Output:
["PASS", "FAIL", "PASS", "FAIL"]


```python
def check_API_Response(status_code):

   if status_code == 200:
      return "PASS"
   else:
       return "FAIL"


#Input
responses = [200, 404, 200, 500]

output = []

for r_status in responses:
    output.append(check_API_Response(r_status))

print(output)
```

Q5 — Timeout Failure Detection

If response time > 500ms, mark it "SLOW" else "OK"

Input:
[120, 560, 230, 900]

Output:
["OK", "SLOW", "OK", "SLOW"]


```python
def Check_Response_Time(Time):

  if Time > 500:
    return "SLOW"

  else:
    return "OK"

#Input
Response_Time = [120, 560, 230, 900]

output = []

for result in Response_Time:
  output.append(Check_Response_Time(result))

print(output)
```
